### Dataset description:

We are using the truncated version of [kaggle dataset](https://www.kaggle.com/datasets/husainsb/lendingclub-issued-loans/data?select=lc_loan.csv), called "lc_loan.csv", which contains data from a company called Lending Club on all the loans they issued between 2007 and 2015, as well as the status of that loan. The original file is pretty large sitting at 887,380 rows, our truncated version has 180,000 rows and 74 attributes, we are looking to predict the status of each loan (whether it's deliquent or paid in full), and will do using the attribute loan_status.


I don't think it's a good use of time to provide a description of each of the 74 attributes in this dataset. Here is a table of 10 attributes we will use in our model

### Implementation/technical merit

Loans with the status "Current" or "Fully Paid" make up 91% of the data, since we are more interested in accurately identifying loans with problems (I.E. deliquent) we will probably have to oversample problematic loans when developing our training data. 

We also have a large amount of attributes (70+), and will only use a portion of them for our model. A couple methods to pare down the number of attributes are:
1) Use correlation matrices to keep only one representative attribute from groups of highly correlated columns.
2) Drop obviously irrelevant attributes, like ID's and descriptions of loan/occupation
3) Train a basic model, evaluate it's performance, and then drop attributes that have minimal impact on model accuracy at predicting loan_status

### Potential impact of the results
A model that accurately predicts loan deliquency allows lenders to take steps to mitigate their losses, avoid risk, and save money. The stakeholders that would be interested in these results would be banks, financial institutions, and private equity the lends money to individuals.
